### Load Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import imageio

plt.style.use('bmh')
sns.set(rc={'figure.figsize':(9,6)})
```

### Data Preprocessing

### Import Data

##### Population

```python
population = pd.read_excel('../data/gapminder_population.xlsx',index_col=0)
population.head()
```

```python
population.isna().sum()
```

```python
population.columns
```

```python
# rename population index name
population.index.name='country'
population.head()
```

```python
# convert the population table into long format
population.reset_index(inplace=True)
population_long = population.melt(id_vars = 'country', var_name = 'year', value_name = 'population')
population_long.head()
```

##### Life Expectancy

```python
life_exp = pd.read_excel('../data/gapminder_lifeexpectancy.xlsx',index_col=0)
life_exp.head()
```

```python
life_exp.isna().sum()
```

```python
life_exp.columns
```

```python
# rename life exp index name
life_exp.index.name='country'
life_exp.head()
```

```python
# convert the life expectancy table into long format
life_exp.reset_index(inplace=True)
life_exp_long = life_exp.melt(id_vars = 'country', var_name = 'year', value_name = 'life_expectancy')
life_exp_long.head()
```

##### Fertility

```python
fertility = pd.read_csv('../data/gapminder_total_fertility.csv',index_col=0)
fertility.head()
```

```python
fertility.isna().sum()
```

```python
fertility.columns
```

```python
# convert fertility columns to int type
fertility.columns = fertility.columns.astype(int)
fertility.columns
```

```python
# rename fertility index name
fertility.index.name='country'
fertility.head()
```

```python
# convert the fertility table into long format
fertility.reset_index(inplace=True)
fertility_long = fertility.melt(id_vars = 'country', var_name = 'year', value_name = 'fertility_rate')
fertility_long.head()
```

##### Continents

```python
continents = pd.read_csv('../data/continents.csv',sep=';')
continents.head()
```

### Merge Tables

```python
df = pd.merge(fertility_long, life_exp_long, how = 'outer', on = ['country','year'])
df.head()
```

```python
df = pd.merge(df,population_long, how='outer',on= ['country','year'] )
df.head()
```

```python
df = pd.merge(df,continents,how ='inner',on='country')
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

#### General Analysis

create an animated scatterplot 

```python
# plot and save correlation between life expectancy and fertility for years between 1960-2015
for year in range(1960,2016):
    subset = df.loc[df['year'] == year]
    ax = sns.scatterplot(data=subset,x='life_expectancy',y='fertility_rate', hue='continent',
                        alpha=0.6,size =subset['population']//10_000,sizes=(10,1000))
    
    handles, labels  =  ax.get_legend_handles_labels()
    plt.axis((0,100,0,10))
    plt.title("Correlation between life expentency and fertility rate for year = %s" % str(year))
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility Rate')
    ax.legend(handles,labels[0:7],loc = 'center left',bbox_to_anchor=(1, 0.5),fontsize=13)
    plt.savefig('../figures/life_exp/life_exp_%s.png' %str(year),bbox_inches='tight')
    plt.close()
```

```python
# generate gif file
images = []

for i in range(1960, 2016):
    filename = '../figures/life_exp/life_exp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('../figures/animation.gif', images, fps=10)
```

### Visualizing Using Bokeh

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, CategoricalColorMapper
from bokeh.palettes import Spectral6
from bokeh.models import Slider
from bokeh.layouts import column, row
from bokeh.io import curdoc

output_notebook()
```

plot correlation between life expectancy and fertility for 2000

```python
year = 2000

subset = df.loc[df['year'] == year]

fertility_rate = subset['fertility_rate']
life_expectancy = subset['life_expectancy']

# Create the figure: p
p = figure(title = f'Correlation between life expentency and fertility rate for year {year}',plot_height=600,plot_width=900, x_axis_label='life_expectancy', y_axis_label='fertility_rate')

# Add a circle glyph to the figure p
p.circle(life_expectancy, fertility_rate)


# Display the plot
show(p)
```

make the plot fancier by considering the size of bubbles regarding the population and colorizing bubbles regarding the continent

```python
year = 2000

subset = df.loc[df['year'] == year]
subset['population'] = subset['population']//15000000

source = ColumnDataSource(subset)

# Define collormapper
mapper = CategoricalColorMapper( factors = list(subset['continent'].unique()),
                                 palette=Spectral6)

# Create the figure: p
p = figure(title = f'Correlation between life expentency and fertility rate for year {year}',plot_height=600,plot_width=900, x_axis_label='life_expectancy', y_axis_label='fertility_rate')



# Add a circle glyph to the figure p
p.circle('life_expectancy', 'fertility_rate', size = 'population',fill_alpha=0.6,
         source=source,color={'field': 'continent', 'transform': mapper},legend_field='continent')
p.legend.location = 'bottom_left'


# Display the plot
show(p)
```
